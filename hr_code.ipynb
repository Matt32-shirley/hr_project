# HR ATTRITION ANALYSIS 

In this analysis, we aid an HR Departement by helping them know what factors contribute to attrition. This data is presented through breakdown structures, graphs, and company information. The data that was used to determine these results was employee information: monthly income, age, years at company, gender, etc. 
The data was broken down into segments that will help the HR Department know wher to focuss their efforts to reduce attrition.

### Split Data Inot Training and Test Sets. 

```r
set.seed(3456)
trainIndex <- createDataPartition(hr_data $ Attrition, p = .8, 
                                  list = FALSE, time = 1)
hr_train <- hr_data [trainIndex,]
hr_test <- hr_data [-trainIndex,]
```

### First Records In HR Train Data Set

```r
head(hr_train)
```

### Descriptive Calculations For Each Field in Data Base

```r
summary(hr_train)
```

### Prediction Attriton Based on all Variables. 

```r
Attrition_Model_1 <- glm(Attrition ~., data=hr_train)
```

```r
summary(Attrition_Model_1)
```

### Prediction of Attrition Variables Based on Attrition_Model_1

```r

Attrition_Model_2 <- glm(Attrition~Age + Gender + MaritalStatus + JobRole 
    + OverTime + EducationField +BusinessTravel + DistanceFromHome + JobInvolvement +
            JobSatisfaction + NumCompaniesWorked + YearsSinceLastPromotion, data = hr_train)
```

```r
summary(Attrition_Model_2)
```

### Prediction of Attrition Variables Based on Attrition_Model_2

```r
Attrition_Model_3 <- glm(Attrition~Age + Gender + MaritalStatus + JobRole + OverTime
                        + BusinessTravel + DistanceFromHome + JobInvolvement + JobSatisfaction
                        + NumCompaniesWorked, data=hr_train)
```

```r
summary(Attrition_Model_3)
```

### Predict Attrition

```r
varImp(Attrition_Model_3)
```

```r
hr_test$Attrition_pred <- predict(Attrition_Model_2, newdata = hr_test, type = "response")
```

```r
head(hr_test)
```

### Write the File to an Excel File

```r
write.csv(hr_test,"attrition_HR.csv", row.names = FALSE)
```
