### Split data inot training and test sets. 

```r
set.seed(3456)
trainIndex <- createDataPartition(hr_data $ Attrition, p = .8, 
                                  list = FALSE, time = 1)
hr_train <- hr_data [trainIndex,]
hr_test <- hr_data [-trainIndex,]
```

### Prepare the train data to predict the attrition rate. To figure out the significant variable that will help you solve for you the AUC (area under the curve) the closer your model is to 1 the better. This will be preformed on all three models to find out what variable are most significant. 

```r
#Model 1
Attrition_Model.1 <- glm(Attrition ~., data=hr_train)
```

```r
summary(Attrition_Model.1)
```

### Prediction of Attrition Variables Based on Attrition_Model_1

```r
#make prediction using all variable
hr_test$attrition_prob_lr_ALL <- predict(Attrition_Model.1, newdata=hr_test,type="response")
```

```r
head(hr_test)
```

### AUC = .796 based of significant varialble from model 1.

```r
ROC_1<-roc(hr_test$Attrition, hr_test$attrition_prob_lr_ALL)
auc(ROC_1)
plot(ROC_1, print.auc=TRUE, main="Original")
```

### After solving for significant varible from models 1, 2, & 3 I compared ROC charts to figure our what variable model has the most significant varables. Based on the graph below Model 1 has more significant variable than other models.  

```r
#evaluate your prediction model
ROC <-roc(hr_test$Attrition, hr_test$Attrition_prob_lr)

plot(ROC, print.auc=TRUE,col="green")
plot(ROC_1,print.auc.y=.4,print.auc=TRUE,col="blue",add=TRUE,Axes=TRUE)
```

```r
#build model that predict the attrition
hr_train_model_1<-glm(Attrition~.,data=hr_train)
summary(hr_train_model_1)

hr_test$Attrition_prediction <- predict(hr_train_model_1, newdata = hr_test,type='response')
head(hr_test)
```

### Below are three trees, all include significant variables. The purpose of each tree is to prioritze each variable based on how signicant the variable is. Each tree has a different number of variable. 

```r
# Generate the decision tree (some variables).
shh(hr_Attrition.1tree <- randomForest(Attrition ~Age + JobInvolvement+NumCompaniesWorked+
                               YearsInCurrentRole+YearsSinceLastPromotion 
                               +MaritalStatus, data = hr_train, importance=TRUE))
varImpPlot(hr_Attrition.1tree)
summary(hr_Attrition.1tree)
```

```r
# Generate the decision tree (some variables).
shh(hr_Attrition.2tree <- randomForest(Attrition ~Age +JobRole+ JobInvolvement+NumCompaniesWorked+YearsInCurrentRole+ 
                     YearsSinceLastPromotion +MaritalStatus, data = hr_train, importance=TRUE))
varImpPlot(hr_Attrition.2tree)
summary(hr_Attrition.2tree)
```

```r
# Generate the decision tree (all variables).
shh(hr_Attrition.3tree <- randomForest(Attrition~Age + Gender + MaritalStatus + JobRole + OverTime
                        + BusinessTravel  + JobInvolvement +YearsInCurrentRole
                        + NumCompaniesWorked, data=hr_train,importance=TRUE))
varImpPlot(hr_Attrition.3tree)
summary(hr_Attrition.3tree)
```

### Below are three predictions that are based on the trees above. They include significant data from each tree. 

### Tree 1

```r
#write new column into test data with my predictions from the firs model with all variables
hr_test$attrition_tree_pred_some1<-predict(hr_Attrition.1tree, newdata=hr_test,type="response")
head(hr_test)
```

### Tree 2

```r
#write new column into test data with my predictions from the second model with all variables
hr_test$attrition_tree_pred_some2<-predict(hr_Attrition.2tree, newdata=hr_test,type="response")
head(hr_test)
```

### Tree 3

```r
#write new column into test data with my predictions from the firs model with all variables
hr_test$attrition_tree_pred_all<-predict(hr_Attrition.3tree, newdata=hr_test,type="response")
head(hr_test)
```

### This ROC was generated based on significant data from trees 1 & 3. The area under the curve for tree 1 .782 making it the most out of the two, because its nuber is closer to one. This means there are more significant variables in that tree. Tree 3's area under the curve is .733

```r
ROC <-roc(hr_test$Attrition, hr_test$Attrition_prob_lr)
plot(ROC, print.auc=TRUE,col="green")

ROC_2 <-roc(hr_test$Attrition, hr_test$attrition_tree_pred_all)
plot(ROC_2, print.auc = TRUE, col= "blue", print.auc.y=.4, add = TRUE) 
auc(ROC)
```

### After reviewing the data and analyzing multiple factors to prevent attrition. It is fair to say that a majority of the company’s attrition is coming from those employees who are, under 40 years of age. A reason for this could be that they are payed less, work more overtime, along with other outlying factors that we don’t have data on. 
    When doing research on why this maybe I read tow interesting articles. One being from business.dailypay.com and the other from towardsdatascience.com. Both on the topic of people 40 years and younger (millennial generation) being so quick to move from job to job. Both article included information that HR could benefit from based off of the data I evaluated. 
    In the article written by business.dailypay.com they said,
           “Only 29 percent of millennials are engaged at work, and 60 percent are open to a new job opportunity. This lack of engagement could be seen as a response to the disconnect between workers and managers. Many employers expect engaged employees but do not necessarily incentivize that engagement, and millennial dissatisfaction with their current position may be the effect of management."
     After reading this along with the article found on towardsdatascience.com which stated, 
        “Though everyone’s individual motive is different, there is plenty of research to indicate the variety of reasons millennials hop jobs as often as they do. They:
            •	Are disengaged a work
            •	Don’t have a compelling reason to stay
            •	Feel their company offers too few perks
            •	Want a better work/life balance"
    A key area for HR to focus on could be training their managers on how to engage with their employees. This would build a bond and a loyalty that would make employees feel important and noticed. Another solution would be to invest in their employees early in their careers. Make them feel important and valued. This to will lead to trust and loyalty resulting in less attrition.
